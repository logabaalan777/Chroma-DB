```python
import os
import PyPDF2
import pdfplumber

def split_pdf_and_extract_text(input_pdf, output_dir):
    # List to store extracted text from each page
    extracted_texts = []

    # Open the input PDF file
    with open(input_pdf, 'rb') as pdf_file:
        # Create a PdfFileReader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        # Loop through each page
        for page_num in range(num_pages):
            # Create a PdfFileWriter object for each page
            pdf_writer = PyPDF2.PdfWriter()
            pdf_writer.add_page(pdf_reader.pages[page_num])

            # Output file name
            output_pdf_filename = os.path.join(output_dir, f'page_{page_num + 1}.pdf')

            # Write the page to a new PDF file
            with open(output_pdf_filename, 'wb') as output_pdf:
                pdf_writer.write(output_pdf)

            print(f'Created: {output_pdf_filename}')

            # Extract text from the page and save it to the list
            text = extract_text_from_pdf(output_pdf_filename)
            extracted_texts.append({"content": text})

    return extracted_texts

def extract_text_from_pdf(pdf_filename):
    # Open the input PDF file
    with pdfplumber.open(pdf_filename) as pdf:
        # There should be only one page in this PDF
        page = pdf.pages[0]
        text = page.extract_text()
        return text

# Example usage
input_pdf = '/content/HDFC-Life-Sanchay-Plus-Life-Long-Income-Option-101N134V19-Policy-Document.pdf'
output_dir = '/content'
extracted_texts = split_pdf_and_extract_text(input_pdf, output_dir)

# Print extracted texts
for idx, page_dict in enumerate(extracted_texts):
    print(f'Page {idx + 1}:\n{page_dict["content"]}\n')
```

```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
embedding_model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def embed_text(texts):
    inputs = tokenizer(texts, padding = True, truncation = True, return_tensors = "pt")
    outputs = embedding_model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim = 1)
    return embeddings.detach().numpy()
```

```python
document_texts = [doc['content'] for doc in extracted_texts]
```

```python
document_embeddings = embed_text(document_texts)
```
